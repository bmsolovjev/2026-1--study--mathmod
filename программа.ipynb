### Закон Снеллиуса
Закон преломления определяет связь между углами падения и преломления:
$$n_1 \sin(\theta_i) = n_2 \sin(\theta_t)$$

**Вывод сохранения тангенциальной компоненты волнового вектора:**

1.  **Определение волнового числа:** Волновое число в среде $j$ равно $k_j = k_0 n_j$, где $k_0 = 2\pi/\lambda$ — волновое число в вакууме, а $n_j$ — показатель преломления среды.

2.  **Тангенциальная компонента волнового вектора:** Если угол $\theta$ измеряется относительно нормали к поверхности (ось X), то тангенциальная (параллельная поверхности, ось Z) компонента волнового вектора $k_z$ дается как:
  $$k_z = k \sin(\theta) = (k_0 n) \sin(\theta)$$

3.  **Применение к падающей и преломленной волнам:**
    *   Для падающей волны в среде 1:
        $$k_{iz} = k_0 n_1 \sin(\theta_i)$$
    *   Для преломленной волны в среде 2:
        $$k_{tz} = k_0 n_2 \sin(\theta_t)$$

4.  **Связь с Законом Снеллиуса:** Подставим выражения для $k_{iz}$ и $k_{tz}$ в Закон Снеллиуса $n_1 \sin(\theta_i) = n_2 \sin(\theta_t)$:
  $$\frac{k_{iz}}{k_0} = \frac{k_{tz}}{k_0}$$
  Отсюда следует:
  $$k_{iz} = k_{tz}$$

5.  **Сохранение для отраженной волны:** Аналогично, из условий отражения (угол падения равен углу отражения $\theta_i = \theta_r$) и того, что отраженная волна находится в той же среде $n_1$, получаем:
  $$k_{rz} = k_0 n_1 \sin(\theta_r) = k_0 n_1 \sin(\theta_i) = k_{iz}$$

Таким образом, в векторной форме это выражается через сохранение тангенциальной (параллельной границе) компоненты волнового вектора:
$$k_{iz} = k_{rz} = k_{tz}$$
где в нашей системе координат граница — $x=0$, а тангенциальная ось — $z$.

### Формулы Френеля (для s-поляризации)
Амплитудные коэффициенты отражения ($R$) и пропускания ($T$) для волны, где вектор поля перпендикулярен плоскости падения:

1. **Коэффициент отражения:**
$$r = \frac{E_r}{E_i} = \frac{k_{ix} - k_{tx}}{k_{ix} + k_{tx}}$$

2. **Коэффициент пропускания:**
$$t = \frac{E_t}{E_i} = \frac{2k_{ix}}{k_{ix} + k_{tx}}$$

Здесь $k_{ix}$ и $k_{tx}$ — проекции волновых векторов на нормаль к границе (ось $x$).

### Цитата из М. Борн, Э. Вольф «Основы оптики»

> «**Законы отражения и преломления.** Если на границу двух однородных сред с разными оптическими свойствами падает плоская волна, она разделяется на две волны: проходящую во вторую среду и отраженную. Существование двух волн вытекает из граничных условий, так как легко видеть, что последние невозможно удовлетворить, если не постулировать наличия как проходящей, так и отраженной волн... Плоскость, определяемая вектором $s^{(i)}$ и нормалью к границе, называется **плоскостью падения**. Соотношения показывают, что и $s^{(r)}$, и $s^{(t)}$ лежат в этой плоскости.»

> «**Формулы Френеля.** Разложим каждый вектор на компоненты — параллельную (индекс $||$) и перпендикулярную (индекс $\perp$) плоскости падения... Волны этих двух типов независимы друг от друга.»

**Формулы Френеля в общем виде:**
    Приведенные ранее формулы Френеля для s-поляризации (электрическое поле перпендикулярно плоскости падения) выглядят так:
    $$r_s = \frac{n_1 \cos\theta_i - n_2 \cos\theta_t}{n_1 \cos\theta_i + n_2 \cos\theta_t}$$ (или в терминах $k_x$: $r = \frac{k_{ix} - k_{tx}}{k_{ix} + k_{tx}}$)
    $$t_s = \frac{2n_1 \cos\theta_i}{n_1 \cos\theta_i + n_2 \cos\theta_t}$$ (или в терминах $k_x$: $t = \frac{2k_{ix}}{k_{ix} + k_{tx}}$)  

### Поляризация?



*   `lambda_val`: Длина волны света (в метрах), цвет.
*   `n1`: Показатель преломления первой среды.
*   `n2`: Показатель преломления второй среды.
*   `theta_deg`: Угол падения света в градусах. Это угол между направлением падающей волны и нормалью к поверхности.
*   `A_I`: Амплитуда падающей волны. Это значение, которое определяет интенсивность света перед взаимодействием с поверхностью.


*   **`k0` (Волновое число в вакууме):**
    *   Это базовое волновое число, которое описывает распространение света в вакууме. Оно указывает, сколько радианов фазы волны приходится на единицу длины в вакууме.
    *   **Формула:** `k0 = 2 * π / λ`
        *   где `λ` (`lambda_val` в коде) — это длина волны света в вакууме.

*   **`k1` (Волновое число в первой среде):**
    *   Это волновое число света, когда он распространяется в первой оптической среде (с показателем преломления `n1`).
    *   **Формула:** `k1 = k0 * n1`

*   **`k2` (Волновое число во второй среде):**
    *   это волновое число света, когда он распространяется во второй оптической среде (с показателем преломления `n2`).
    *   **Формула:** `k2 = k0 * n2`

    

```python
import numpy as np

def wave_fields(lambda_val, n1, n2, theta_deg, A_I):
    """
    Расчет отраженных и прошедших полей на границе x=0.
    x < 0: среда 1 (n1) - свет падает отсюда
    x > 0: среда 2 (n2) - свет проходит сюда
    Положительная ось X направлена вниз (в среду 2).
    Положительная ось Z направлена вправо.
    """
    theta_rad = np.radians(theta_deg)
    k0 = 2 * np.pi / lambda_val
    k1 = k0 * n1
    k2 = k0 * n2

    # Падающий луч k_I (в плоскости xz, y=0)
    # При x=0 границе, k_Ix направлен к границе (в положительную x)
    k_Ix = k1 * np.cos(theta_rad)
    k_Iy = 0.0
    k_Iz = k1 * np.sin(theta_rad)

    k_I = np.array([k_Ix, k_Iy, k_Iz])

    # Закон Снеллиуса (k_z)
    k_Rz = k_Iz
    k_Tz = k_Iz

    # Отраженный луч k_R (x < 0, движется в отрицательную x)
    k_Rx = -k_Ix # Отражение: x-компонента меняет знак
    k_R = np.array([k_Rx, 0.0, k_Rz])

    # Прошедший луч k_T (x > 0)
    # k_Tx^2 + k_Tz^2 = k2^2
    k_Tx_sq = k2**2 - k_Tz**2
    if k_Tx_sq < 0:
        k_Tx = 1j * np.sqrt(-k_Tx_sq) # Полное внутреннее отражение, затухание в +x
    else:
        k_Tx = np.sqrt(k_Tx_sq) # Положительный, так как идет в x > 0

    k_T = np.array([k_Tx, 0.0, k_Tz])

    # Коэффициент отражения для нормального падения
    r_normal = (n1 - n2) / (n1 + n2)
    # Коэффициент преломления для нормального падения
    t_normal = (2 * n1) / (n1 + n2)

    R_amp = A_I * r_normal
    T_amp = A_I * t_normal

    return {
        "k_I": k_I, "k_R": k_R, "k_T": k_T,
        "A_I": A_I, "R": R_amp, "T": T_amp
    }

# Параметры
lam = 500e-9  # Длина волны
n1 = 1.0      # Воздух
n2 = 1.5      # Стекло
theta = 30.0  # Угол падения в градусах (0 - нормальное падение)
A_init = 1.0  # Начальная амплитуда

res = wave_fields(lam, n1, n2, theta, A_init)


print(f"Угол падения: {theta}°")
print(f"k_I (Падающий): {res['k_I']}")
print(f"k_R (Отраженный): {res['k_R']}")
print(f"k_T (Преломлённый): {res['k_T']}")
print(f"Амплитуда R (по нормальному падению): {res['R']:.4f}")
print(f"Амплитуда T (по нормальному падению): {res['T']:.4f}")
```
